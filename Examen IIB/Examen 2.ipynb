### 14) Crear un Dataframe con 10 registros y 3 columnas, A B C, llenas de números randomicos del 1 al 10 enteros. Crear una nueva columna con el calculo por fila (A * B ) / C

```python
arr_trece = np.random.randint(1, 10, 30).reshape(10, 3)
df_quince = pd.DataFrame(arr_trece, columns = ['A', 'B', 'C'])
resultados = []
for index in df_quince.index:
    resultados.append((df_quince['A'][index] * df_quince['B'][index]) / df_quince['C'][index] )

df_quince['Resultados'] = resultados
df_quince
```
